```python
import plotly.graph_objects as go
import numpy as np
```

```python
# Make data.
x = np.linspace(0, 1, 100)
y = np.linspace(0,1, 100)
xx, yy = np.meshgrid(x, y)
```

```python
zz = np.divide(xx, (xx + yy), out=np.zeros_like(xx), where=(xx!=0)|(yy!=0))
```

```python
fig = go.Figure(
        go.Surface(
            x=x,
            y=y,
            z=zz,
        )
    )

camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=-1.25, y=-1.25, z=1.25)
)

fig.update_layout(
    autosize=True,
    width=1000,
    height=1000,
    scene_camera=camera,
)
```
